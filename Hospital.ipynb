# Data Requirements

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
derm=pd.read_csv("documents/dermatology.txt",header=None)
derm.columns=('erythema','scaling','borders','itching','koebner','polygonal','follicular','oral mucosal','knee and elbow','scalp','family history','melanin','eosinophils','PNL','fibrosis','exocytosis','acanthosis','hyperkeratosis','parakeratosis','clubbing','elongation','thinning','spongiform','munro','focal','disappearance','vacuolisation','spongiosis','saw-tooth','horn plug','perifollicular','monoluclear infiltrate','band-like infiltrate','age','ok')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split 
from sklearn import metrics 
```

```python
train,test=train_test_split(derm,test_size=0.10)
train.head()
```

```python
X_train=train[['polygonal','oral mucosal','focal','vacuolisation','saw-tooth','band-like infiltrate']]
```

```python
y_train=train.melanin
```

```python
X_test=test[['polygonal','oral mucosal','focal','vacuolisation','saw-tooth','band-like infiltrate']]
y_test=test.melanin
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
prediction=model.predict(X_test)
```

```python
Line=prediction.astype(int)
Line
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model=KNeighborsRegressor()
model.fit(X_train,y_train)
prediction=model.predict(X_test)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model=KNeighborsClassifier()
model.fit(X_train,y_train)
prediction=model.predict(X_test)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
