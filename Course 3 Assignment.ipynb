# Data Understanding and Preparation

- Our objective is to find what factors affect melanin so below is how I approach that.

```python
plt.figure(figsize=(20,15))
sns.heatmap(derm.corr(),annot=True)
```

- What we have here is a covariance matrix. What this means is that all these numbers that you see are the relationship between two variables. I want you to find melanin on the y-axis. I then want you to look horizontally and find the six numbers that are close to one. When you do, correspond those numbers with the names on the x-axis. The closer the covariance is to one, then that means the factor is most related.

- Once you followed the instructions above, you probably realized that through this covariance matrix, we have determined the six factors that affect melanin. They are polygonal, oral mucosal , focal , vacuolisation , saw-tooth, band-like infiltrate. Just in case that's not enough proof:

```python
sns.lmplot('melanin','polygonal',derm)
sns.lmplot('melanin','focal',derm)
```

- The couple line plots above is just to show you that the factors that have close correlation have linear relationships which prove the point. Now, it's time for the last step.

# Modeling and Evaluation

- Now that we have determined the factors, we shall construct multiple models and figure out which model gives the best accuracy so we can compare new values to the original.

```python
from sklearn.cross_validation import train_test_split 
from sklearn import metrics 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

- These are all imported tools using sklearn which is powerful for machine learning. Now, you may be wondering why I am only picking regression algorithims and not classification. The reason why is because I want a continuous array of values as soon as the final model has been picked. I'm not looking for something discrete or specific.

```python
train,test=train_test_split(derm,test_size=0.10)
train.head()
```

- What you see here is a splitting of my training and test data. I'm splitting that to 90 percent training and 10 percent test data. Now, let's create the X and y train and test variables.

```python
X_train=train[['polygonal','oral mucosal','focal','vacuolisation','saw-tooth','band-like infiltrate']]
y_train=train.melanin
```

```python
X_test=test[['polygonal','oral mucosal','focal','vacuolisation','saw-tooth','band-like infiltrate']]
y_test=test.melanin
```

- Now let's make models and find percentages for each.

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model=KNeighborsRegressor()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model=RandomForestRegressor()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model=GradientBoostingRegressor()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model=BaggingRegressor()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model=DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

- So these are all the percentages for all six models and from what you see, a majority of them had high training percentages, but test score percentages drastically went down. The most consistent model out of all them would be the Linear Regression, so let's make an array of values using the Linear Regression Model, so we can compare the new one with the original.

```python
model=LinearRegression()
model.fit(X_train,y_train)
prediction=model.predict(X_test)
```
