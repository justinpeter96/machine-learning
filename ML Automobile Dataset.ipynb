```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import seaborn as sns
auto=pd.read_csv('imports-85.data.txt',header=None)
auto.columns=['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style',
             'drive-wheels','engine-location','wheel-base','length','width','height',
             'curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system',
             'bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
auto 
```

```python
auto.isnull().sum()
```

```python
auto.info
```

```python
a=auto[auto['normalized-losses']!='?']
b=(a['normalized-losses'].astype(int)).mean()
auto['normalized-losses']=auto['normalized-losses'].replace('?',b).astype(int)
```

```python
a=auto[auto['body-style']=='sedan']
a['num-of-doors'].value_counts()
```

```python
a=auto['num-of-doors'].map({'two':2,'four':4,'?':4})
auto['num-of-doors']=a
```

```python
a=auto[auto['price']!='?']
b=(a['price'].astype(int)).mean()
auto['price']=auto['price'].replace('?',b).astype(int)
```

```python
a=auto[auto['horsepower']!='?']
b=(a['horsepower'].astype(int)).mean()
auto['horsepower']=auto['horsepower'].replace('?',b).astype(int)
```

```python
a=auto[auto['bore']!='?']
b=(a['bore'].astype(float)).mean()
auto['bore']=auto['bore'].replace('?',b).astype(float)
```

```python
a=auto[auto['stroke']!='?']
b=(a['stroke'].astype(float)).mean()
auto['stroke']=auto['stroke'].replace('?',b).astype(float)
```

```python
a=auto[auto['peak-rpm']!='?']
b=(a['peak-rpm'].astype(float)).mean()
auto['peak-rpm']=auto['peak-rpm'].replace('?',b).astype(float)
```

```python
a=auto['num-of-cylinders'].map({'four':4,'five':5,'six':6,'?':4})
auto['num-of-doors']=a
```

```python
print('Car makers in the data set are',auto['make'].unique())
```

```python
fig = plt.figure(figsize=(15, 10))
ax=sns.countplot(auto['make'],palette='dark',edgecolor='k',linewidth=2)
plt.xticks(rotation='vertical')
plt.xlabel('Car Maker',fontsize=20)
plt.ylabel('Number of Cars',fontsize=20)
plt.title('Cars count by manufacturer',fontsize=30)
ax.tick_params(labelsize=15)
#plt.yticks(rotation='vertical')
plt.show()
```

```python
print('Different types of cars',auto['body-style'].unique())
```

```python
fig = plt.figure(figsize=(15, 10))
cars_type=auto.groupby(['body-style']).count()['make']
ax=cars_type.sort_values(ascending=False).plot.bar(edgecolor='k',linewidth=2)
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.15, p.get_height()+1),fontsize=11)
plt.xticks(rotation='vertical')
plt.xlabel('Body Type',fontsize=20)
plt.ylabel('Number of Cars',fontsize=20)
plt.title('Count of Cars by Body Type',fontsize=30)
ax.tick_params(labelsize=15)
#plt.yticks(rotation='vertical')
plt.show()
```

```python
a=auto.groupby(['make','body-style']).count().reset_index()
a=a[['make','body-style','symboling']]
a.columns=['make','style','count']
a=a.pivot('make','style','count')
a.dropna(thresh=3).plot.barh(width=0.85)
plt.show()
```

```python
plt.figure(1)
plt.subplot(221)
ax1=auto['engine-type'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='orange',edgecolor='k',linewidth=2)
plt.title("Number of Engine Type frequency diagram")
plt.ylabel('Number of Engine Type',fontsize=15)
ax1.tick_params(labelsize=15)
plt.xlabel('engine-type',fontsize=15);


plt.subplot(222)
ax2=auto['num-of-doors'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='purple',edgecolor='k',linewidth=2)
plt.title("Number of Door frequetncy diagram")
plt.ylabel('Number of Doors',fontsize=15)
ax2.tick_params(labelsize=15)
plt.xlabel('num-of-doors',fontsize=15);

plt.subplot(223)
ax3=auto['fuel-type'].value_counts(normalize= True).plot(figsize=(10,8),kind='bar',color='green',edgecolor='k',linewidth=2)
plt.title("Number of Fuel Type frequency diagram")
plt.ylabel('Number of vehicles',fontsize=15)
plt.xlabel('fuel-type',fontsize=15)
ax2.tick_params(labelsize=15)

plt.subplot(224)
ax4=auto['body-style'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='red',edgecolor='k',linewidth=2)
plt.title("Number of Body Style frequency diagram")
plt.ylabel('Number of vehicles',fontsize=15)
plt.xlabel('body-style',fontsize=15);
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(15, 8))
mileage=auto.groupby(['make']).mean()
mileage['avg-mpg']=((mileage['city-mpg']+mileage['highway-mpg'])/2)
ax=mileage['avg-mpg'].sort_values(ascending=False).plot.bar(edgecolor='k',linewidth=2)
plt.xticks(rotation='vertical')
plt.xlabel('Car Maker',fontsize=20)
plt.ylabel('Number of Cars',fontsize=20)
plt.title('Fuel Economy of Car Makers',fontsize=30)
ax.tick_params(labelsize=20)
#plt.yticks(rotation='vertical')
plt.show()
plt.show()
```
